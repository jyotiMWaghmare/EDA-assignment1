```python
cat_col=['fuel-type', 'aspiration','num-of-doors', 'body-style', 'drive-wheels']
i=0
while i < 5:
    fig = plt.figure(figsize=[20,5])  
    
    
    plt.subplot(1,3,1)
    sns.countplot(x=cat_col[i], data=df)
    i += 1
    
    
    plt.subplot(1,3,2)
    sns.countplot(x=cat_col[i], data=df)
    i += 1
    
      
    plt.show()
```

```python
num_col=['length', 'width', 'height', 'curb-weight']
i=0
while i < 4:
    fig = plt.figure(figsize=[20,5])  
    
    
    plt.subplot(1,2,1)
    sns.boxplot(x=num_col[i], data=df)
    i += 1
    
    
    plt.subplot(1,2,2)
    sns.boxplot(x=num_col[i], data=df)
    i += 1
    
      
    plt.show()
```
